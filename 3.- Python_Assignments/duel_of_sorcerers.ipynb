```python
# DUEL OF SORCERERS

# TASK 1:
gandalf = [10, 11, 13, 30, 22, 11, 10, 33, 22, 22]
saruman = [23, 66, 12, 43, 12, 10, 44, 23, 12, 17]
spells = (gandalf, saruman)

# TASK 2:
gandalf_wins = 0
saruman_wins = 0

# TASK 3:
for spells in zip(gandalf, saruman):
    gandalf_spell = spells[0]
    saruman_spell = spells[1]
    if gandalf_spell > saruman_spell:
        gandalf_wins = gandalf_wins + 1
    elif gandalf_spell < saruman_spell:
        saruman_wins = saruman_wins + 1

# TASK 4:
if gandalf_wins > saruman_wins:
    print('Task Battle: Gandalf wins.')
elif gandalf_wins < saruman_wins:
    print('Task Battle: Saruman wins.')
elif gandalf_wins == saruman_wins:
    print('Task Battle: Tie')
    
# BONUS 1:
POWER = {
    'Fireball': 50, 
    'Lightning bolt': 40, 
    'Magic arrow': 10, 
    'Black Tentacles': 25, 
    'Contagion': 45
}
gandalf = ['Fireball', 'Lightning bolt', 'Lightning bolt', 'Magic arrow', 'Fireball', 
           'Magic arrow', 'Lightning bolt', 'Fireball', 'Fireball', 'Fireball']
saruman = ['Contagion', 'Contagion', 'Black Tentacles', 'Fireball', 'Black Tentacles', 
           'Lightning bolt', 'Magic arrow', 'Contagion', 'Magic arrow', 'Magic arrow']
spells = (gandalf, saruman)

# BONUS 2:
gandalf_wins = 0
saruman_wins = 0

# BONUS 3:
gandalf_power = []
for value in gandalf:
    gandalf_power.append(POWER[value])
saruman_power = []
for value in saruman:
    saruman_power.append(POWER[value])
    
# BONUS 4:
round_winner = None
for spell_power in zip(gandalf_power, saruman_power):
    gandalf_power = spell_power[0]
    saruman_power = spell_power[1]
    if gandalf_power > saruman_power:
        if round_winner != 'Gandalf':
            round_winner = 'Gandalf'
            gandalf_wins = 1
        else:
            gandalf_wins = gandalf_wins + 1
    elif gandalf_power < saruman_power:
        if round_winner != 'Saruman':
            round_winner = 'Saruman'
            saruman_wins = 1
        else:
            saruman_wins = saruman_wins + 1
    elif gandalf_power == saruman_power:
        round_winner = None
    if gandalf_wins == 3:
        print('Bonus Battle: Gandalf wins.')
        break
    elif saruman_wins == 3:
        print('Bonus Battle: Saruman wins.')
        break

# BONUS 5:
gandalf_power = []
for value in gandalf:
    gandalf_power.append(POWER[value])
gandalf_average_power = sum(gandalf_power) / len(gandalf_power)
print("The avarage power of Gandalf's spells in Bonus Battle is", gandalf_average_power)

saruman_power = []
for value in saruman:
    saruman_power.append(POWER[value])
saruman_average_power = sum(saruman_power) / len(saruman_power)
print("The avarage power of Saruman's spells in Bonus Battle is", saruman_average_power)

# BONUS 6:
sorcerers_average_power = sum(gandalf_power + saruman_power) / len(gandalf_power + saruman_power)
print("The avarage power of both sorcerers spells in Bonus Battle is", sorcerers_average_power)

gandalf_stdev = sum([(gandalf_average_power - value) for value in (gandalf_power + saruman_power)]) / len(gandalf_power + saruman_power)
print("The standard deviation of Gandalf power with respect to the avarage power of both sorcerers in Bonus Battle is", gandalf_stdev)

saruman_stdev = sum([(saruman_average_power - value) for value in (gandalf_power + saruman_power)]) / len(gandalf_power + saruman_power)
print("The standard deviation of Saruman power with respect to the avarage power of both sorcerers in Bonus Battle is", saruman_stdev)
```
