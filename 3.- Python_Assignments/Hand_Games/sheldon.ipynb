```python
while True:
    print()
    n_rounds = input('Enter the number of rounds do you want to play. It must be an odd number equal or below 9: ')
    print()
    try:
        n_rounds = int(n_rounds)
        if n_rounds % 2 == 0:
            print('This is not an odd number. Try again.')
            continue
        elif n_rounds > 9:
            print('This is a number above 9. Try again.')
            continue
        else:
            break
    except ValueError:
        print('This is not a valid enter. Try again.')
        continue
print('You selected to play', n_rounds, 'rounds.')
print()
rspls_text = """The rules of this game, created by Sam Kass and Karen Bryla, are:
- Sissors cuts Paper
- Paper covers Rock
- Rock crushes Lizard
- Lizard poisons Spock
- Spock smashes Scissors
- Scissors decapitates Lizard
- Lizard eats Paper
- Paper disaproves Spock
- Spock vaporizes Rock
- Rock crushes Scissors
"""
print(rspls_text)

import math
rounds_to_win = math.ceil(n_rounds/2)
print('You have to beat the CPU', rounds_to_win, 'times to win the game.')
print()

def round_winner():
    gestures = ['rock!','paper!','scissors!', 'lizard!', 'spock!']
    from random import randint
    cpu_move = randint(0, (len(gestures) - 1))
    while True:
        player_move = (input('Make your move. Remember, it must be rock (r), paper (p), scissors (s), lizardc(l) or spock (sp): '))
        try:
            if player_move == 'r':
                player_move = 'rock!'
                break
            elif player_move == 'p':
                player_move = 'paper!'
                break
            elif player_move == 's':
                player_move = 'scissors!'
                break
            elif player_move == 'l':
                player_move = 'lizard!'
                break
            elif player_move == 'sp':
                player_move = 'spock!'
                break
        except player_move != 'r' or 'p' or 's' or 'l' or 'sp':
            continue
    print('You played...', player_move)
    print('CPU played...', gestures[cpu_move])

    matrix = [[0, 1, 2, 1, 2],
             [2, 0, 1, 2, 1],
             [1, 2, 0, 2, 1],
             [1, 2, 1, 0, 2],
             [2, 1, 2, 1, 0]]
    conv = {'rock!':0, 'paper!':1, 'scissors!': 2, 'lizard!':3, 'spock!':4}
    results = [0, 1, 2]
    winner = (results[matrix[conv[player_move]][cpu_move]])

    if winner == 0:
        print("It's a TIE!")
        print()
        return(int(0))
    elif winner == 1:
        print('YOU LOSE!')
        print()
        return(int(1))
    elif winner == 2:
        print('YOU WIN!')
        print()
        return(int(2))

x_rounds = 0
tie_count = 0
cpu_score = 0
player_score = 0
while x_rounds != n_rounds:
    x_rounds += 1
    print('Round NÂº:', x_rounds, 'of', n_rounds)
    winner_id = round_winner()
    if winner_id == 0:
        tie_count += 1
        print('SCOREBOARD')
        print('Tie count =', tie_count)
        print('CPU =', cpu_score, 'wins')
        print('Player =', player_score, 'wins')
        print()
    elif winner_id == 1:
        cpu_score += 1
        print('SCOREBOARD')
        print('Tie count =', tie_count)
        print('CPU =', cpu_score, 'wins')
        print('Player =', player_score, 'wins')
        print()
    elif winner_id == 2:
        player_score +=1
        print('SCOREBOARD')
        print('Tie count =', tie_count)
        print('CPU =', cpu_score, 'wins')
        print('Player =', player_score, 'wins')
        print()

    if n_rounds == x_rounds and cpu_score == player_score:
        print ("Technology is important, but the only thing that really matters is what we do with it. We have tied.")
        print()
        break
    if rounds_to_win == cpu_score or n_rounds == x_rounds and cpu_score > player_score:
        print("To date, no computer that has been designed is aware of what it is doing, but most of the time, humans aren't either. I'm sorry, you lost.")
        print()
        break
    if rounds_to_win == player_score or n_rounds == x_rounds and player_score > cpu_score:
        print("One machine can do the work of fifty ordinary humans, but no machine can do the work of an extraordinary one. Congratulations, you won. ")
        print()
        break
```
