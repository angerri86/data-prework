```python
# ROBIN HOOD

# TASK 1:
points = [(4, 5), (-0, 2), (4, 7), (1, -3), (3, -2), (4, 5), (3, 2), (5, 7), (-5, 7), (2, 2), (-4, 5), (0, -2),
          (-4, 7), (-1, 3), (-3, 2), (-4, -5), (-3, 2), (5, 7), (5, 7), (2, 2), (9, 9), (-8, -9)]

double_hits = list(set([hits for hits in points if points.count(hits) > 1]))

print("TASK 1: The points where an arrow hits another arrow are:", (double_hits))

# TASK 2:
Q1 = []
Q2 = []
Q3 = []
Q4 = []
axis = []
arrows = 0

while arrows < len(points):
    for x, y in points[arrows : arrows + 1]:
        if x > 0 and y > 0:
            Q1.append(points[arrows])
        elif x > 0 and y < 0:
            Q2.append(points[arrows])
        elif x < 0 and y > 0:
            Q3.append(points[arrows])
        elif x < 0 and y < 0:
            Q4.append(points[arrows])
        elif x == 0 and y == 0:
            axis.append(points[arrows])
        arrows += 1
Q1_hits = (len(Q1))
Q2_hits = (len(Q2))
Q3_hits = (len(Q3))
Q4_hits = (len(Q4))
axis_hits = (len(axis))

print('TASK 2: Arrows that hit in Q1 =', Q1_hits)
print('TASK 2: Arrows that hit in Q2 =', Q2_hits)
print('TASK 2: Arrows that hit in Q3 =', Q3_hits)
print('TASK 2: Arrows that hit in Q4 =', Q4_hits)
print('TASK 2: Arrows that hit in axis =', axis_hits)

# TASK 3:
#-> Distance from center
import math
def center_dist(hit):
    x = hit[0]
    y = hit[1]
    distance = math.sqrt(x**2 + y**2)
    return distance

#-> Distances list
distances = []
for hits in points:
    distances.append(round(center_dist(hits), 2))

#-> Closest points list
marks = 0
best_mark = []
for hit in distances:
    if hit == min(distances):
        best_mark.append(points[marks])
    marks += 1
    
print("TASK 3: Closest points to the center =", best_mark)
print("TASK 3: Closest distance to the center =", min(distances))

# TASK 4:
#-> No target hits
target_radius = 9
sum(1 for target_hit in distances if target_hit > 9)
print("TASK 4: Arrows that didn't hit the target =", sum(1 for target_hit in distances if target_hit > 9))

# TASK 5:
#-> Radial stripes hits
radius = 9
stripe = 1
arrows = 1
while stripe <= radius:
    for hit in distances:
        if hit <= stripe and hit > stripe-1:
            arrows += 1
    if arrows == 1:
        print('TASK 5:', arrows, 'arrow hit stripe', stripe)
    elif arrows > 1:
        print('TASK 5:', arrows, 'arrows hit stripe', stripe)
    arrows = 0
    stripe += 1
            
        
```
